```python
def f(x):
    return x**2
  
  
def g(x):
    return x**(1/2)
  
  
x = np.linspace(0, 2, 1000)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.fill_between(x, f(x), g(x), where=[(x > 0) and (x < 1) for x in x])
plt.show()
def aso(n, x):
    return quad(f(x), 1, np.inf, args=(n, x))
```

```python
x = sy.symbols('x', real=True)
f = sy.sin(x)**4 * sy.exp(-7*x)
sy.integrate(f, x)
```

```python
a, b = sy.symbols('a b', real=True, positive=True)
f = sy.cos(b*x)* sy.exp(-a*x)

sy.integrate(f, x).simplify()

```

```python
f = (1+sy.sqrt(x))**sy.Rational(1,3) / sy.sqrt(x)
sy.integrate(f, x).simplify()
```

```python
f = sy.exp(x) / sy.sqrt(sy.exp(2*x) + 9)
sy.integrate(f, (x, 0, sy.log(4)))
```

```python
f = 16*sy.atan(x) / (1+x**2)
sy.integrate(f, (x, 0, sy.oo))
```
